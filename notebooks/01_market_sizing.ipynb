# Market sizing - Indian EV two-wheeler

Quick TAM / SAM / SOM cross-check.
Baseline numbers from SIAM FY24 + Vahan dashboard. Framing the market as a 5-year cumulative addressable window (FY24-FY28).


```python
import pandas as pd

df = pd.read_csv('../data/market_data_fy24.csv')
df
```

## Step 1 - project EV unit sales over 5 years

Growth tapers off year by year as the base grows. Industry consensus is ~30%+ CAGR in the near term, slowing as scale arrives.


```python
fy24_ev_units = 1_011_000
yoy_growth = [1.50, 1.40, 1.35, 1.30]  # tapering growth multipliers FY25 to FY28

ev_units = [fy24_ev_units]
for g in yoy_growth:
    ev_units.append(ev_units[-1] * g)

for i, u in enumerate(ev_units):
    print(f'FY{24+i}: {u/1e6:.2f}M units')

total_units_5yr = sum(ev_units)
print(f'\nCumulative FY24-FY28: {total_units_5yr/1e6:.2f}M units')
```

## Step 2 - TAM

Avg price assumption: Rs 80k over the 5-year period. Lower than today's ~Rs 100k because mass-market scaling and FAME-II phase-out push prices down. CRISIL projects sub-Rs 80k average by FY27.


```python
avg_price_5yr = 80_000  # Rs per unit, average over FY24-FY28
tam_5yr = total_units_5yr * avg_price_5yr
print(f'TAM (5-yr cumulative): Rs {tam_5yr/1e7:,.0f} Cr')
```

*Round to ~Rs 85k Cr for the headline number - lots of uncertainty in the 5-year average price.*


## Step 3 - SAM

Filter: urban districts (Vahan: 78% of EV sales), and the Rs 80k-1.5L band where new entrants are most likely to compete.


```python
urban_share = 0.78  # Vahan FY24
target_band_share = 0.55  # share of urban EV sales in Rs 80k-1.5L band (JMK 2024)

sam_5yr = tam_5yr * urban_share * target_band_share
print(f'SAM (5-yr): Rs {sam_5yr/1e7:,.0f} Cr')
```

## Step 4 - SOM

Target: 3% share of SAM, over 5 metros. Cross-check against Ather - Ather hit Rs 1,789 Cr in FY24 at ~9% share, suggesting today's SAM is roughly Rs 20k Cr. That tracks with our FY27 SAM projection of Rs 38k Cr (about 2x growth in 4 years, consistent with EV adoption curve).


```python
som_share = 0.03
som_5yr = sam_5yr * som_share
print(f'SOM target (5-yr cumulative): Rs {som_5yr/1e7:,.0f} Cr')
```

## Sensitivity - EV-share assumption is the swing variable

What if EV penetration is lower than the 18% consensus by FY27?


```python
for share in [0.12, 0.15, 0.18, 0.21]:
    # rough adjustment: scale cumulative units linearly by share ratio
    adj_units = total_units_5yr * (share / 0.18)
    adj_tam = adj_units * avg_price_5yr
    adj_sam = adj_tam * urban_share * target_band_share
    adj_som = adj_sam * som_share
    print(f'Penetration {share:.0%} -> TAM Rs {adj_tam/1e7:,.0f} Cr | SAM Rs {adj_sam/1e7:,.0f} Cr | SOM Rs {adj_som/1e7:,.0f} Cr')
```

## Takeaway

- **TAM**: ~Rs 85-90k Cr 5-yr cumulative at 18% penetration (Rs 60-105k Cr range across sensitivities)
- **SAM**: ~Rs 38k Cr urban + target band
- **SOM**: ~Rs 1,140 Cr at 3% share over 5 metros
